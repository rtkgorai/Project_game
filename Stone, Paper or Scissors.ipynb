```python
import random
def game(c,y):
    if c==y:
        return 0
    elif c=='s':
        if y=='x':
            return -1
        elif y=='p':
            return 1
    elif c=='p':
        if y=='s':
            return -1
        elif y=='x':
            return 1
    else:
        if y=='p':
            return -1
        elif y=='s':
            return 1

    
    
if __name__ == "__main__":
    print("Welcome to the Stone, Paper or Scissors Game!")
    print()
    print('''The rules of the "Stone, Paper or Scissors" game are as simple as a cup of tea:
              The game determines the winner based on predefined rules: 
              stone beats scissors, scissors beats paper, and paper beats stone.''')
    print()
    s=0 #player score
    so=0 #computer score  
    g=None
    while(g!='n'):
        print("****************************************************")
        y=input("Choose stone(s), paper(p), or scissors(x):")
        r=random.randint(1,3)
        if(r==1):
            c='s'
        elif(r==2):
            c='p'
        else:
            c='x'
        if c=='s':
            print("Computer chose:stone(s)")
        elif c=='s':
            print("Computer chose:paper(p)")
        else:
            print("Computer chose:scissors(x)")
        print()
        if(game(c,y)==1):
            print("You win this round!")
            s=s+1
        elif(game(c,y)==-1):
            print("Computer wins this round!")
            so=so+1
        else:
            print("It's a tie!")


        print("****************************************************")
        print("Want to play the game again?")
        g=input("Type yes(y) or no(n):")
        
        

    print("****************************************************")
    print("Final Score:")   
    print("Player:",s)
    print("Computer:",so)
    print()
    if(s>so):
        print("Congratulations! You've win the game")
    elif(so==s):
        print("OOPS!tie")
    else:
        print("You've lose the game")
    print("****************************************************")
```
